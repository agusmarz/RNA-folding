```python
x = df['base_pairs'][0]
import ast
import numpy as np
big_n = 80
def pass_str_to_array(x):
    x = ast.literal_eval(x)


    output = np.zeros((big_n*big_n,), dtype=int)
    for i in x:
        multiplicador = i[0] -1
        resto = i[1] -1
        output[resto+multiplicador*big_n] = 1

    return output
df['base_pairs'] = df['base_pairs'].apply(pass_str_to_array)

df
```

```python
import torch as tr

NT_DICT = {
    "R": ["G", "A"],
    "Y": ["C", "U"],
    "K": ["G", "U"],
    "M": ["A", "C"],
    "S": ["G", "C"],
    "W": ["A", "U"],
    "B": ["G", "U", "C"],
    "D": ["G", "A", "U"],
    "H": ["A", "C", "U"],
    "V": ["G", "C", "A"],
    "N": ["G", "A", "C", "U"],
}

VOCABULARY = ["A", "C", "G", "U"]


class OneHotEmbedding:
    def __init__(self):
        self.pad_token = "-"
        self.vocabulary = VOCABULARY
        self.emb_size = len(self.vocabulary)

    def seq2emb(self, seq, pad_token="-"):

        seq = seq.upper().replace("T", "U")
        emb_size = len(VOCABULARY)
         # ACA CAMBIE EL LARGO DE TODOS LOS EMBEDDINGS PARA QUE SIEMPRE SEAN big_n !!!!!!!
        emb = tr.zeros((emb_size, big_n), dtype=tr.float)

        for k, nt in enumerate(seq):
            if nt == pad_token:
                continue
            if nt in VOCABULARY:
                emb[VOCABULARY.index(nt), k] = 1
            elif nt in NT_DICT:
                v = 1 / len(NT_DICT[nt])
                ind = [VOCABULARY.index(n) for n in NT_DICT[nt]]
                emb[ind, k] = v
            else:
                raise ValueError(f"Unrecognized nucleotide {nt}")

        return emb
embedding = OneHotEmbedding()


df['embedding'] = df['sequence'].apply(embedding.seq2emb)

print(df)
df["embedding"][0].shape
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset

X = np.array(df["embedding"].tolist())
y = np.array(df["base_pairs"].tolist())


print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=1)


X_train = torch.tensor(X_train, dtype=torch.float32)
X_val = torch.tensor(X_val, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
y_val = torch.tensor(y_val, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)


batch_size = 4

train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

val_dataset = TensorDataset(X_val, y_val)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

test_dataset = TensorDataset(X_test, y_test)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class BinaryOutputMLP(nn.Module):
    def __init__(self):
        super(BinaryOutputMLP, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(4 * big_n, 3096)
        self.fc2 = nn.Linear(3096, 4096)
        self.fc3 = nn.Linear(4096, big_n*big_n)
        self.verbose = False

    def forward(self, x):
        x = self.flatten(x)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

    def predict(self, x):
        with torch.no_grad():
            x = self.forward(x)
            return (torch.sigmoid(x) > 0.5).float()  # necesito que todos los output sean o ceros o unos
model = BinaryOutputMLP()


```

```python
from torch.cuda.amp import autocast, GradScaler
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 10
model = model.to(device)
criterion = nn.BCEWithLogitsLoss()
scaler = GradScaler()

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for batch_idx, (inputs, targets) in enumerate(train_loader):
        inputs = inputs.to(device)
        targets = targets.to(device)

        with autocast():
            outputs = model(inputs)
            loss = criterion(outputs, targets)

        optimizer.zero_grad()
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()

        del inputs, targets, outputs, loss
        torch.cuda.empty_cache()

    print(f"Epoch [{epoch+1}/{num_epochs}], Training Loss: {running_loss/len(train_loader)}")

```

```python
import torch
import torch.nn as nn
from sklearn.metrics import f1_score
from tqdm import tqdm

def f1_triangular(ref, pred):
    """Compute F1 from the upper triangular connection matrix"""
    # get upper triangular matrix without diagonal
    ref = ref.reshape(80, 80)
    pred = pred.reshape(80, 80)
    ind = torch.triu_indices(ref.shape[0], ref.shape[1], offset=1)
    ref = ref[ind[0], ind[1]].cpu().numpy().ravel()
    pred = pred[ind[0], ind[1]].cpu().numpy().ravel()
    return f1_score(ref, pred, zero_division=0)

def contact_f1(ref_batch, pred_batch, lengths, th=0.5, reduce=True, method="triangular"):
    """Compute F1 from base pairs. Input goes to sigmoid and then thresholded"""
    f1_list = []

    if isinstance(ref_batch, float) or len(ref_batch.shape) < 3:
        ref_batch = [ref_batch]
        pred_batch = [pred_batch]
        lengths = [lengths]

    for ref, pred, l in zip(ref_batch, pred_batch, lengths):
        # ignore padding
        ind = tr.where(ref != -1)
        pred = pred[ind].view(-1)
        ref = ref[ind].view(-1)

        # pred goes from -inf to inf
        pred = tr.sigmoid(pred)
        pred[pred <= th] = 0

        if method == "triangular":
            f1 = f1_triangular(ref, pred > th) # Solamente vamos a usar este!!! (creo)
        elif method == "shift":
            ref = mat2bp(ref)
            pred = mat2bp(pred)
            _, _, f1 = f1_shift(ref, pred)
        else:
            raise NotImplementedError

        f1_list.append(f1)

    if reduce:
        return tr.tensor(f1_list).mean().item()
    else:
        return tr.tensor(f1_list)
def mcc(pred, ref):
    """pred, ref are 2D binary matrices"""
    tp = np.logical_and(pred, ref).sum()
    fp = pred.sum() - tp
    fn = ref.sum() - tp
    tn = np.logical_and(np.logical_not(pred), np.logical_not(ref)).sum()
    mcc = (tp * tn - fp * fn) / np.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn) + 1e-8)
    return mcc


def postprocessing(preds, masks):
    """Postprocessing function using viable pairing mask.
    Inputs are batches of size [B, N, N]"""
    print(preds.shape)
    print(masks.shape)

    if masks is not None:
        preds = preds.multiply(masks)

    y_pred_mask_triu = tr.triu(preds)
    y_pred_mask_max = tr.zeros_like(preds)
    for k in range(preds.shape[0]):
        y_pred_mask_max_aux = tr.zeros_like(y_pred_mask_triu[k, :, :])

        val, ind = y_pred_mask_triu[k, :, :].max(dim=0)
        y_pred_mask_max[k, ind[val > 0], val > 0] = val[val > 0]

        val, ind = y_pred_mask_max[k, :, :].max(dim=1)
        y_pred_mask_max_aux[val > 0, ind[val > 0]] = val[val > 0]

        ind = tr.where(y_pred_mask_max[k, :, :] != y_pred_mask_max_aux)
        y_pred_mask_max[k, ind[0], ind[1]] = 0

        y_pred_mask_max[k] = tr.triu(y_pred_mask_max[k]) + tr.triu(
            y_pred_mask_max[k]
        ).transpose(0, 1)
    return y_pred_mask_max
def evaluate_rna_predictor(model, test_loader, criterion, device):
    model.eval()
    metrics = {"loss": 0, "f1": 0, "f1_post": 0}

    loader = test_loader

    if hasattr(model, 'verbose') and model.verbose:
        loader = tqdm(loader)

    with torch.no_grad():
        for batch in loader:
            inputs, targets = batch

            lengths = [int(targets.size(1))] * targets.size(0)

            inputs = inputs.to(device)
            targets = targets.to(device)

            y_pred = model(inputs)
            loss = criterion(y_pred, targets)
            metrics["loss"] += loss.item()

            y_pred = y_pred.reshape(y_pred.size(0), 80, 80)
            targets = targets.reshape(targets.size(0), 80, 80)

            canonical_mask = torch.ones_like(targets).to(device)
            y_pred_post = postprocessing(y_pred, canonical_mask)

            f1 = contact_f1(targets.cpu(), y_pred.cpu(), lengths, th=0.5, reduce=True, method="triangular")
            f1_post = contact_f1(targets.cpu(), y_pred_post.cpu(), lengths, th=0.5, reduce=True, method="triangular")

            metrics["f1"] += f1
            metrics["f1_post"] += f1_post

    for k in metrics:
        metrics[k] /= len(loader)

    if hasattr(model, 'scheduler_name') and model.scheduler_name == "plateau":
        model.scheduler.step(metrics["f1_post"])

    return metrics

metrics = evaluate_rna_predictor(model, test_loader, criterion, device)
print(metrics)
```
